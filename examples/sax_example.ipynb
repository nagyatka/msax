## Imports

```python
import numpy as np
import matplotlib.pyplot as plt

from msax.msax import normalize, paa, sax, breakpoints
```

```python
# Creating a random time series
mu, sigma = 0, 0.1
x = np.random.normal(mu, sigma, 100)

fig = plt.figure(figsize=(15,4))
plt.plot(x)
```

## SAX

The SAX method involves the following steps:
    1. Normalization (normalize function)
    2. PAA transformation (paa function)
    3. Discretization (alphabetize function)

These steps are included in the sax function, which executes the original SAX transformation. The required parameters:
 - x: input array
 - w: window size for PAA
 - a: alphabet size for discretization 
 
The resulting array contains the indices of symbols from the alphabet. Based on indices, min_dist function can determine the distance between two symbolic representation really fast.

```python
x_sax = sax(x, w=2, a=20)
x_sax
```

The transformation result could be visualized too.

```python
fig = plt.figure(figsize=(15,4))
plt.plot(x_sax)
```
